<h2>Importing Libraries</h2>

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
import missingno as msno
import warnings
warnings.simplefilter("ignore")
```

```python
pd.set_option('display.max_columns',85)
pd.set_option('display.max_rows',85)
```

<h2>Importing and Exploring Data</h2>


```python
airbnb=pd.read_csv('Airbnb_Open_Data.csv')
```

```python
airbnb
```

```python
airbnb.info()
```

```python
airbnb.isna().count().count()
```

```python
msno.matrix(airbnb)
plt.show()
```

<h2>Drop unnecessary columns</h2>

```python
airbnb.drop(['id','NAME','license','host id','house_rules','host name'],axis=1,inplace=True)
```

```python
airbnb
```

<h2>Imputing and removing Null values</h2>

<h3>First we have to check our columns again after dropping unnecessary of them</h3>

```python
msno.matrix(airbnb)
plt.show()
```

<h3>Now we'il impute nulls our categorical columns</h3>

<h4>1. host_identify_vertified column</h4>

```python
airbnb['host_identity_verified'].value_counts()
```

```python
airbnb['host_identity_verified'].fillna('unconfirmed',inplace=True)
```

<h4>2. Country column</h4>

```python
airbnb['country'].value_counts()
```

```python
airbnb['country'].fillna('United States',inplace=True)
```

<h3>Now we'il impute nulls our numerical columns but we need to fix a categorical columns that it should've been a numerical</h3>

<h4>1. Price column</h4>

```python
airbnb["price"] = airbnb["price"].str.replace("$", "")
```

```python
airbnb["price"] = airbnb["price"].str.replace(",", "")
```

```python
airbnb["price"]=pd.to_numeric(airbnb['price'],errors='coerce')
```

<h4>2. Service fee column</h4>

```python
airbnb["service fee"] = airbnb["service fee"].str.replace("$", "")
airbnb["service fee"] = airbnb["service fee"].str.replace(",", "")
```

```python
airbnb["service fee"]=pd.to_numeric(airbnb["service fee"], errors='coerce')
```

```python
msno.matrix(airbnb)
plt.show()
```

<h2>Now Let's remove and impute nulls</h2>

<h3>1. Remove nulls</h3>

```python
airbnb.dropna(subset=['service fee','country code','instant_bookable','cancellation_policy'
                      ,'price','minimum nights','Construction year','review rate number','availability 365'
                     ,'calculated host listings count','neighbourhood group'],inplace=True)
```

```python
msno.matrix(airbnb)
plt.show()
```

<h3>2. Impute nulls</h3>

```python
airbnb.fillna('unreviewed',inplace=True)
```

```python
msno.matrix(airbnb)
plt.show()
```

```python
airbnb.reset_index(inplace=True)
```

<h2>Finally our data are clean Let's go to our last stage and it's Visualizing our data</h2>

```python
plt.figure(figsize=(20,26))
airbnb = airbnb[(airbnb['minimum nights'] > 0) & (airbnb['minimum nights'] <= 366)]
sns.countplot(data=airbnb,x='minimum nights')
plt.xticks(rotation=90)

plt.show()
```
